```python
estimate_charges=(30,50,100)
```

```python
ages= non_smokers_df.age
ages
```

```python
def estimate_charges(ages,w,b):
    return w*ages + b
```

```python
estimated_charges= estimate_charges(ages,w,b)
```

```python
plt.plot(ages,estimated_charges,'r');
sns.scatterplot(data=non_smokers_df,x='age',y='charges',alpha=0.7,s=15)
plt.xlabel('Age')
plt.ylabel('Estimated Charges')
plt.show()
```

```python
targets=non_smokers_df.charges
```

```python
plt.plot(ages,estimated_charges,'r',alpha=0.9);
plt.scatter(ages,targets,s=8,alpha=0.8);
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend(['Estimate','Actual'])
```

```python
def try_parameters(w,b):
    ages=non_smokers_df.age
    targets=non_smokers_df.charges
    estimated_charges=estimate_charges(ages,w,b)
    plt.plot(ages,estimated_charges,'r',alpha=0.9);
    plt.scatter(ages,targets,s=8,alpha=0.8);
    plt.xlabel('Age')
    plt.ylabel('Charges')
    plt.legend(['Estimate','Actual'])
```

```python
try_parameters(60,100)
try_parameters(400,5000)
try_parameters(350,3000)
```
